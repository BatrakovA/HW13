Будем парсить страницу со свежеми новостям на habr.com/ru/all/.

Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

KEYWORDS = ['python', 'парсинг']

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

В итоге должен формироваться датафрейм вида: <дата> - <заголовок> - <ссылка>

```python
import requests
```

```python
keywords = ['python', 'парсинг']
```

```python
import time
```

```python
import pandas as pd
```

```python
URL = 'https://habr.com/ru/all/'
```

```python
req = requests.get(URL)
time.sleep(0.3)
print(req.text)
```

```python
from bs4 import BeautifulSoup
```

```python
params = {
    'q': 'python'
    }
req = requests.get(URL, params)
time.sleep(0.3)
soup = BeautifulSoup(req.text)
news_blocks = soup.find_all('article', class_='tm-articles-list__item')
news_blocks
```

```python
soup.find('span', class_='tm-article-snippet__datetime-published').find('time').get('title')
```

```python
soup.find('h2', class_='tm-article-snippet__title').find('span').get_text()
```

```python
soup.find('a', class_='tm-article-snippet__title-link').get('href')
```

```python
for e in keywords:
    params = {
    'q': e
    }
    req = requests.get(URL, params)
    time.sleep(0.3)
    soup = BeautifulSoup(req.text)
    news_blocks = soup.find_all('article', class_='tm-articles-list__item')
    articles_date = list(map(lambda x: x.find('span', class_='tm-article-snippet__datetime-published').find('time').get('title'), news_blocks))
    articles_text = list(map(lambda x: x.find('h2', class_='tm-article-snippet__title').find('span').get_text(), news_blocks))
    print(articles_date + articles_text)
```
